```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, svm, neighbors
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('image-train.csv')
df1 = pd.read_csv('image-test.csv')
df.head()
```

```python
for i in df.columns:
    df[i] = df[i].astype(np.float64)
for i in df1.columns:
    df1[i] = df1[i].astype(np.float64)
df['label'] = df['label'].astype(int)
```

```python
x = np.array(df.drop(['label'],1))
x = preprocessing.scale(x)
y = np.array(df.label)
```

```python
clf = neighbors.KNeighborsClassifier()
clf = RandomForestClassifier(max_depth=8, random_state=0)
print('starting fitting')
clf.fit(x,y)
print('done fitting')
print(clf.score(x,y))

# clf = KMeans(n_clusters=10)
# print('starting fitting')
# clf.fit(x,y)
# print('done fitting')
# print(clf.score(x,y))
```

```python
predict_x = np.array(df1)
predict_x = predict_x.reshape(len(predict_x),-1)
prediction = clf.predict(predict_x)
```

```python
import csv
with open('upload_image.csv','w') as fp:
    sw = csv.writer(fp, delimiter=',')
    sw.writerow(['ImageId','Label'])
    for i in range(len(prediction)):
        sw.writerow([i+1,int(prediction[i])])
```
